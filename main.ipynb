```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from joypy import joyplot
file_path= "csv/2018 FIFA World Cup Squads.csv"
df = pd.read_csv(file_path)
```

```python
#Biểu đồ phân bố tuổi của các cầu thủ
plt.figure(figsize=(10, 6))
joyplot(
    data=df[["Age", "Team"]].pivot(columns="Team", values="Age"),
    overlap=0.5,
    colormap=sns.color_palette("coolwarm", as_cmap=True)
)

# Thêm tiêu đề
plt.title("Phân bố độ tuổi cầu thủ theo đội tuyển")
plt.xlabel("Tuổi")
plt.savefig('pic1.png')
plt.show()
```

```python
#Biểu đồ số bàn thắng theo đội
goals_by_team = df.groupby("Team")["Goals"].sum().sort_values(ascending=False)
goals_by_team.plot(kind="barh", color="skyblue", width=0.7)

plt.title("Tổng số bàn thắng theo đội")
plt.xlabel("Số bàn thắng")
plt.ylabel("Đội")
plt.gca().invert_yaxis()

plt.savefig('pic2.png')
plt.show()
```

```python
top_scorers = df.sort_values(by="Goals", ascending=False).head(20)  # Lấy top 20 cầu thủ

# Tạo biểu đồ
plt.figure(figsize=(12, 8))
plt.barh(top_scorers["Player"], top_scorers["Goals"], color="skyblue")
plt.title("Top 20 Cầu Thủ Ghi Nhiều Bàn Thắng Nhất (World Cup 2018)", fontsize=16)
plt.xlabel("Số Bàn Thắng", fontsize=12)
plt.ylabel("Cầu Thủ", fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis="x", linestyle="--", alpha=0.7)

# Hiển thị biểu đồ
plt.tight_layout()
plt.savefig('pic3.png')
plt.show()
```

```python
forwards = df[df["Position"] == "FW"]
plt.figure(figsize=(10, 6))
sns.regplot(x=forwards["Caps"], y=forwards["Goals"], scatter_kws={"color": "red", "alpha": 0.6, "edgecolor": "black"}, line_kws={"color": "blue"})
plt.title("Mối Tương Quan Giữa Số Lần Khoác Áo Đội Tuyển (Caps) và Số Bàn Thắng (Goals) - Tiền Đạo", fontsize=16)
plt.xlabel("Số Lần Khoác Áo Đội Tuyển (Caps)", fontsize=12)
plt.ylabel("Số Bàn Thắng (Goals)", fontsize=12)
plt.grid(linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig('pic4.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(x=forwards["Caps"], y=forwards["Goals"], scatter_kws={"color": "red", "alpha": 0.6, "edgecolor": "black"}, line_kws={"color": "blue"})
plt.title("Mối Tương Quan Giữa Số Lần Khoác Áo Đội Tuyển (Caps) và Số Bàn Thắng (Goals) - Tiền Đạo", fontsize=16)
plt.xlabel("Số Lần Khoác Áo Đội Tuyển (Caps)", fontsize=12)
plt.ylabel("Số Bàn Thắng (Goals)", fontsize=12)
plt.grid(linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig('pic5.png')
plt.show()
```

```python
club_counts = df["Club"].value_counts().head(20)  # Lấy top 20 câu lạc bộ

# Tạo bar chart
plt.figure(figsize=(12, 8))
club_counts.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Top 20 Câu Lạc Bộ Có Nhiều Cầu Thủ Tham Dự World Cup 2018", fontsize=16)
plt.xlabel("Câu Lạc Bộ", fontsize=12)
plt.ylabel("Số Lượng Cầu Thủ", fontsize=12)
plt.xticks(rotation=45, ha="right")  # Xoay nhãn trục x để dễ đọc
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Hiển thị biểu đồ
plt.tight_layout()
plt.savefig('pic6.png')
plt.show()
```
